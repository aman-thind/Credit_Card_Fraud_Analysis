```python
# split normalize data
x = data.drop("Class", axis=1)
y = data.Class
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
# See if both the train and test label distribution are similarly distributed
print("Label distribution y_train")
print(y_train.value_counts(normalize=True))
print("Label distribution y_test")
print(y_test.value_counts(normalize=True))
```

Our dataframe is not balanced, so we will create a sub-sample where the classes are distributed 70/30. With this technique the algorithm can work better but a lot of information is lost that could be useful.

We will use X_train and y_train to do this, reserving the Test to validate .

I will take all the cases of Class=1, and then, to maintain the 70/30 proportion, taking 919 data of Class=0, finally using pd.concat, create the sub_sample (30% = 330, 70%=919)


```python
#First with value counts we see how many values we have from each class
y_train.value_counts()
```

```python
# here we have all fraud cases
X_fraud = data.loc[y_train.loc[y_train == 1].index]
```

```python
# 919 values where Class=0
y_non_fraud = data.loc[y_train.loc[y_train == 0].index][:919]
```

```python
#concat
sub_sample = pd.concat([X_fraud,y_non_fraud],axis=0)
```

```python
#finally a sub_sample with 70/30 proportions
sub_sample.shape
```

```python
new_df_class = sub_sample.Class.value_counts().reset_index()
sns.barplot(x="index",y="Class",data=new_df_class)
sns.despine()
plt.title("Fraud class frequency", size=15)
plt.xlabel("Classes")
plt.ylabel("Frequency")
```

```python
# split normalize data
x_new = sub_sample.drop("Class",axis=1)
y_new = sub_sample.Class
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_new, y_new)
pred_rf=rf.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,pred_rf))
print(accuracy_score(y_test,pred_rf))
print(classification_report(y_test,pred_rf))
```

Random Forest was used in developing fraud detection model to classify a transaction as fraudulent or legitimate. Three metrics were used in evaluating their performances.
Performance improvement could be achieved through developing a fraud detection model using a combination of different algorithms.

Having a data set with non-anonymized features would make this particularly interesting as outputting the feature importance would enable one to see what specific factors are most important for detecting fraudulent transactions.
